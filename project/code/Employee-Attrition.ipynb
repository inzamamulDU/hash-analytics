```python
#merge two dataset for prediction
# Adding both dataset from employees who left with existing employees

df_left['Left']=1

df_not_left['Left']=0

final_df=pd.concat([df_left,df_not_left],axis=0)
```

```python
final_df.head(10)
```

```python
#creating dummy variable
just_dummies = pd.get_dummies(final_df[['dept','salary']])
just_dummies.head(10)
```

```python
#concat two dataframes
final_df=pd.concat([final_df,just_dummies],axis=1)
#drop the 'dept' and''salary' column
final_df.drop(['dept', 'salary'], inplace=True, axis=1)
#show 
final_df.head(10)
```

```python
# Dividing the dataset into X and Y 

X=final_df.drop(['Emp ID','Left'],axis=1)

y=final_df['Left']
```

```python
X
```

```python
y
```

```python
# Splitting the X and y into train and test set

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report

X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=102)
```

```python
DC = DecisionTreeClassifier(max_depth=5, random_state=101, max_features=None, min_samples_leaf=15)
DC.fit(X_train, y_train)
```

```python
y_pred = DC.predict(X_test)
```

```python
#model accuraccy is not so good.
#why the model accuraccy is not so good. let's find out . 
train_accuracy = accuracy_score(y_train, DC.predict(X_train))
test_accuracy = accuracy_score(y_test, y_pred)
print('Train Accuracy:', train_accuracy)
print('Test Accuracy:',test_accuracy)
```

```python
print("employees left is {}% from total dataset".format(round(final_df.Left.value_counts()[1]/len(final_df)*100,2)))
```

```python
#see that 24% only dataset has left employee so the model is imbalance. 
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=10).fit(X_train, y_train)
rfc.fit(X_train, y_train)
```

```python
y_pred_new = rfc.predict(X_test)
```

```python
#A good accuraccy after using random forest classifier but test accuracy is not so good
train_accuracy = accuracy_score(y_train, rfc.predict(X_train))
test_accuracy = accuracy_score(y_test, y_pred_new)
print('Train Accuracy:', train_accuracy)
print('Test Accuracy:',test_accuracy)
```

```python
#doing upsample
```

```python
from sklearn.utils import resample

y_new = final_df['Left']
X_new= final_df.drop(['Left'],axis=1)

X_train, X_test, y_train, y_test = train_test_split(X_new,y_new,train_size=0.75,random_state=50)

X=pd.concat([X_train,y_train],axis=1)
```

```python
# unsampling the minority by adding dummy rows to the left equal to 1 

left_upsampled= resample(X[X.Left==1],replace=True,n_samples=len(X[X.Left==0]),random_state=50)

left_upsampled=pd.concat([X[X.Left==0],left_upsampled])
```

```python
left_upsampled.Left.value_counts()
```

```python
# Preparing for X train and Y train dataset

y_train=left_upsampled.Left
X_train=left_upsampled.drop(['Left'],axis=1)
```

```python
rfc_upsample = RandomForestClassifier(n_estimators=20,random_state = 101)
rfc_upsample.fit(X_train.drop('Emp ID',axis=1), y_train)
```

```python
y_pred_final = rfc_upsample.predict(X_test.drop('Emp ID',axis=1))
```

```python
#model is now at the best position
train_accuracy = accuracy_score(y_train, rfc_upsample.predict(X_train.drop('Emp ID',axis=1)))
test_accuracy = accuracy_score(y_test, y_pred_final)
print('Train Accuracy:', train_accuracy)
print('Test Accuracy:',test_accuracy)
```
